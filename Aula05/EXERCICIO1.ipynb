```python
# Importar bibliotecas
import numpy as np
import matplotlib.pyplot as plt

# Definir parâmetros iniciais
tempo_intervalos = 6  # número de intervalos de tempo
vazao = np.array([10, 20, 30, 5, 35, 40])  # vazão em m³/min

# Definir o volume inicial do tanque e outros parâmetros
volume_inicial = 0  # volume inicial do tanque em m³
volume_tanque = np.zeros(tempo_intervalos + 1)  # Array para armazenar volumes

# Calcular o nível do tanque ao longo do tempo
volume_tanque[0] = volume_inicial
for i in range(tempo_intervalos):
    volume_tanque[i + 1] = volume_tanque[i] + vazao[i]  # Atualizar o volume

# Visualizar os resultados
tempo = np.arange(tempo_intervalos + 1)  # Vetor de tempo

plt.figure(figsize=(10, 6))
plt.plot(tempo, volume_tanque, marker='o', linestyle='-', color='b')
plt.title('Previsão do Nível do Tanque')
plt.xlabel('Intervalos de Tempo')
plt.ylabel('Volume do Tanque (m³)')
plt.grid(True)
plt.xticks(tempo)
plt.show()
```
