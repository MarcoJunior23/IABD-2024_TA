```python
# Importar bibliotecas
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Dados fornecidos
Q = np.array([100, 200, 300, 400, 500, 600, 700]).reshape(-1, 1)  # Quantidade de calor (meltido em coluna)
T = np.array([30, 35, 40, 45, 60, 100, 150])  # Temperatura

# Inicializar o modelo de regressão linear
modelo = LinearRegression()

# Ajustar o modelo aos dados
modelo.fit(Q, T)

# Fazer previsões com os dados fornecidos
T_previsto = modelo.predict(Q)

# Medir o desempenho do modelo
mse = mean_squared_error(T, T_previsto)
r2 = r2_score(T, T_previsto)

print(f"Mean Squared Error: {mse:.2f}")
print(f"R^2 Score: {r2:.2f}")

# Visualizar os resultados
plt.figure(figsize=(10, 6))
plt.scatter(Q, T, color='blue', label='Dados Reais')
plt.plot(Q, T_previsto, color='red', linestyle='-', label='Regressão Linear')
plt.title('Previsão da Temperatura do Forno')
plt.xlabel('Quantidade de Calor (Q)')
plt.ylabel('Temperatura (T)')
plt.legend()
plt.grid(True)
plt.show()

# Fazer uma previsão para novos valores de calor
novos_Q = np.array([150, 250, 350, 450, 550]).reshape(-1, 1)
previsoes = modelo.predict(novos_Q)

print("Previsões para novos valores de calor:")
for q, t in zip(novos_Q.flatten(), previsoes):
    print(f"Quantidade de Calor: {q} -> Temperatura Prevista: {t:.2f}")
```
