```python
# Importar bibliotecas
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Dados fornecidos
Consumo = np.array([50, 140, 30, 75, 123]).reshape(-1, 1)  # Consumo de energia em KW
Demanda = np.array([100, 200, 80, 100, 160])  # Demanda de energia

# Inicializar o modelo de regresso linear
modelo = LinearRegression()

# Ajustar o modelo aos dados
modelo.fit(Consumo, Demanda)

# Fazer previses com os dados fornecidos
Demanda_previsto = modelo.predict(Consumo)

# Medir o desempenho do modelo
mse = mean_squared_error(Demanda, Demanda_previsto)
r2 = r2_score(Demanda, Demanda_previsto)

print(f"Mean Squared Error: {mse:.2f}")
print(f"R^2 Score: {r2:.2f}")

# Visualizar os resultados
plt.figure(figsize=(10, 6))
plt.scatter(Consumo, Demanda, color='blue', label='Dados Reais')
plt.plot(Consumo, Demanda_previsto, color='red', linestyle='-', label='Regresso Linear')
plt.title('Previso da Demanda de Energia com Base no Consumo')
plt.xlabel('Consumo de Energia (KW)')
plt.ylabel('Demanda de Energia')
plt.legend()
plt.grid(True)
plt.show()

# Fazer previses para novos valores de consumo
novos_Consumos = np.array([60, 150, 90, 130]).reshape(-1, 1)
previsoes = modelo.predict(novos_Consumos)

print("Previses para novos valores de consumo:")
for consumo, demanda in zip(novos_Consumos.flatten(), previsoes):
    print(f"Consumo: {consumo} KW -> Demanda Prevista: {demanda:.2f}")
```
