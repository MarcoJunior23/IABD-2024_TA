```python
# Importar bibliotecas
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Dados fornecidos
lojas = ['A', 'B', 'C', 'D']
estoque = [500, 100, 300, 600]
vendas_2022 = {
    'Jan': [100, 80, 150, 50],
    'Fev': [120, 90, 170, 60],
    'Mar': [150, 100, 200, 80],
    'Abr': [180, 110, 230, 90],
    'Mai': [220, 190, 350, 200],
    'Jun': [230, 150, 280, 120],
    'Jul': [250, 170, 300, 140],
    'Ago': [260, 180, 310, 150],
    'Set': [240, 160, 290, 130],
    'Out': [220, 140, 270, 110],
    'Nov': [400, 220, 350, 190],
    'Dez': [300, 350, 400, 250]
}

# Criar o DataFrame com as vendas
df_vendas = pd.DataFrame(vendas_2022, index=lojas)

# Criar a figura e os subplots
fig, axs = plt.subplots(2, 2, figsize=(14, 10), sharex=True, sharey=True)

# Configurar os subplots
meses = df_vendas.columns
for i, loja in enumerate(lojas):
    ax = axs[i // 2, i % 2]
    ax.plot(meses, df_vendas.loc[loja], marker='o', linestyle='-', label=f'Loja {loja}')
    ax.set_title(f'Variação das Vendas - Loja {loja}')
    ax.set_xlabel('Meses')
    ax.set_ylabel('Número de Vendas')
    ax.grid(True)
    ax.legend()
    # Adicionar o estoque como texto no subplot
    ax.text(0.95, 0.05, f'Estoque: {estoque[i]}', transform=ax.transAxes,
            fontsize=12, verticalalignment='bottom', horizontalalignment='right',
            bbox=dict(facecolor='white', alpha=0.5))

# Ajustar layout para não sobrepor os títulos e rótulos
plt.tight_layout()
plt.show()
```
