```python
# Importar bibliotecas
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Dados fornecidos
Renda = np.array([1200, 5000, 8000, 10000, 30000]).reshape(-1, 1)  # Salário em R$
Gastos = np.array([750, 1500, 4000, 6000, 18000])  # Gastos em R$

# Inicializar o modelo de regressão linear
modelo = LinearRegression()

# Ajustar o modelo aos dados
modelo.fit(Renda, Gastos)

# Fazer previsões com os dados fornecidos
Gastos_previsto = modelo.predict(Renda)

# Medir o desempenho do modelo
mse = mean_squared_error(Gastos, Gastos_previsto)
r2 = r2_score(Gastos, Gastos_previsto)

print(f"Mean Squared Error: R${mse:.2f}")
print(f"R^2 Score: {r2:.2f}")

# Visualizar os resultados
plt.figure(figsize=(10, 6))
plt.scatter(Renda, Gastos, color='blue', label='Dados Reais')
plt.plot(Renda, Gastos_previsto, color='red', linestyle='-', label='Regressão Linear')
plt.title('Previsão dos Gastos com Base no Salário')
plt.xlabel('Salário (R$)')
plt.ylabel('Gastos (R$)')
plt.legend()
plt.grid(True)
plt.show()

# Fazer previsões para novos salários
novas_Rendas = np.array([2000, 6000, 12000, 25000]).reshape(-1, 1)
previsoes = modelo.predict(novas_Rendas)

print("Previsões para novos salários:")
for renda, gasto in zip(novas_Rendas.flatten(), previsoes):
    print(f"Salário: R${renda} -> Gastos Previsto: R${gasto:.2f}")
```
