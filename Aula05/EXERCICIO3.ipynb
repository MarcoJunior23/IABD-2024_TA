```python
# Importar bibliotecas
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Dados fornecidos
A = np.array([50, 100, 120, 300]).reshape(-1, 1)  # Área em m² (transformada em coluna)
Valor = np.array([180000, 300000, 375000, 600000])  # Valor do imóvel em R$

# Inicializar o modelo de regressão linear
modelo = LinearRegression()

# Ajustar o modelo aos dados
modelo.fit(A, Valor)

# Fazer previsões com os dados fornecidos
Valor_previsto = modelo.predict(A)

# Medir o desempenho do modelo
mse = mean_squared_error(Valor, Valor_previsto)
r2 = r2_score(Valor, Valor_previsto)

print(f"Mean Squared Error: R${mse:.2f}")
print(f"R^2 Score: {r2:.2f}")

# Visualizar os resultados
plt.figure(figsize=(10, 6))
plt.scatter(A, Valor, color='blue', label='Dados Reais')
plt.plot(A, Valor_previsto, color='red', linestyle='-', label='Regressão Linear')
plt.title('Previsão do Valor do Imóvel com Base na Área')
plt.xlabel('Área (m²)')
plt.ylabel('Valor do Imóvel (R$)')
plt.legend()
plt.grid(True)
plt.show()

# Fazer uma previsão para novas áreas
novas_A = np.array([80, 150, 250, 350]).reshape(-1, 1)
previsoes = modelo.predict(novas_A)

print("Previsões para novas áreas:")
for a, v in zip(novas_A.flatten(), previsoes):
    print(f"Área: {a} m² -> Valor Previsto: R${v:.2f}")
```
