```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Gerando dados
dados_x = np.linspace(0, 10, 10)
dados_y = dados_x * 2  # Exemplo de cálculo para dados_y

# Criando o primeiro gráfico
plt.figure(figsize=(10, 5), facecolor='gray')
ax = plt.axes()
ax.set_facecolor('white')
plt.plot(dados_x, dados_y,
         color='red', linewidth=2,
         marker='o', linestyle='--',
         markeredgecolor='black')
plt.xlabel("Eixo X")  # Adiciona título ao eixo X
plt.ylabel("Eixo Y")  # Corrigido para o eixo Y
plt.title("Exemplo Matplotlib", fontsize=12)

plt.show()  # Exibe o gráfico

# Configurando os subplots
fig, axs = plt.subplots(3, 1, figsize=(12, 12))  # 3 linhas, 1 coluna

# Gráfico 1
axs[0].plot(dados_x, 'r--')
axs[0].set_title("Gráfico 1")

# Gráfico 2
axs[1].plot(dados_x**2, 'b-.')
axs[1].set_title("Gráfico 2")

# Gráfico 3
axs[2].plot(dados_x / 2, 'c-.')
axs[2].set_title("Gráfico 3")

plt.tight_layout()  # Ajusta o layout dos subplots para não sobrepor os títulos
plt.show()  # Exibe os subplots

# Leitura do arquivo CSV
df = pd.read_csv('C:\\Users\\dsadm\\Desktop\\monitoramento_tempo.csv')
print(df.head())  # Exibe as primeiras linhas do DataFrame
```
